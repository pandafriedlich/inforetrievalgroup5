## TASK 2

+ First design the Algorithm to solve the problem in 2-dimension

```python
# Necessary Imports
import numpy as np
import matplotlib.pyplot as plt

# Declaration of Methods
def task2(d, n, rand_min, rand_max):    
    result = 0;
    
    # Prepare the n vectors of dimension n within a random value between rand_min and rand_max
    vectorArray = [[np.random.uniform(rand_min, rand_max) for x in range(d)] for y in range(n)];    
    
    # Calculate list of minimum angle to other vectors
    minAngles = [0] * n;
    for x in range(0, n):
        minAngle = 3.14159;
        for y in range(0, n):
            if(x!=y):
                angleXY = vectorAngle(vectorArray[x],vectorArray[y]);
                if(minAngle>angleXY):
                    minAngle = angleXY;
        minAngles[x] = minAngle;
    
    # Calculate average of minimum angles
    average = arraySum(minAngles)/n;
    result = average;
    return result;

def vectorAngle(v1,v2):     # Returns the angle between two vectors in radians
    return np.arccos(np.dot(v1, v2) /(np.linalg.norm(v1)*np.linalg.norm(v2)));
    
def arraySum(v):            # Returns the sum of all elements
    result = 0;
    dim = len(v);
    for x in range(0, dim):
        result += v[x];
    return result;
    
# Main Algorithm
# Declaration of Parameters

n = 100;                # The number of the vectors in an array
rand_min = -1.0;        # The minimum of the random values for the vector coordinates
rand_max = +1.0;       # The maximum of the random values for the vector coordinates

dimension=2;
result = task2(dimension, n, rand_min, rand_max);
print(result);            # Get the average of the minimum angles
```

+ Repeat the above process for dimension $d=1, \dots, 1000$

```python
dimension_min = 1;      # The dimension of the vectors
dimension_max = 1000;   # The dimension of the vectors

dimensions = list(range(dimension_min, dimension_max+1))
results=[];
for x in range(0, len(dimensions)):
    result = task2(dimensions[x], n, rand_min, rand_max);
    results.append(result);
    results[x]*=180/np.pi
    print(results[x]);       # Get the average of the minimum angles in 1000-dimension(degree)
```

```python
%matplotlib inline
plt.plot(dimensions,results)
plt.title('average minimum angle against the dimensions')
plt.xlabel('dimension')
plt.ylabel('average minangle')
plt.show()
```

+ According to the results and the plot, we can draw the conclusion that for 2 randomly sampled vectors in a d-dimensional space, average minimum angle increases with the increment of the dimensions, when the dimension is close to 1000, average minimum angle is close to 90 degree.

+ To verify the conclusion, we set the sample size as 200 and repeat the above process.

```python
n = 200;  
results2=[];
for x in range(0, len(dimensions)):
    result = task2(dimensions[x], n, rand_min, rand_max);
    results2.append(result);
    results2[x]*=180/np.pi
```

```python
%matplotlib inline
plt.plot(dimensions,results,'k', dimensions,results2, 'r')
plt.legend(('100 samples', '200 samples'))
plt.title('average minimum angle against the dimensions')
plt.xlabel('dimension')
plt.ylabel('average minangle')
plt.show()
```

+ As we can see in the plot, when the sample size increase to 200, the result does not change, namely, the average minimum angle always tend towards 90 degree when the dimension increases.
